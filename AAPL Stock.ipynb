```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tkinter as tk
```

```python
data = pd.read_csv("AAPL.csv") 

prices = data['Close'].values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler()
prices_scaled = scaler.fit_transform(prices)
```

```python
x=prices_scaled[:-1]
y=prices_scaled[1:]
print("X shape:", x.shape, "y shape:", y.shape)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(1,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
print(x.shape, y.shape)
```

```python
model.fit(x,y,epochs=100 ,batch_size=32 , validation_split=0.2)
```

```python
def predict_price():
    
    try:
        days = int(entry.get())  
        current_price = prices_scaled[-1].reshape(1, 1) 
        predictions = []
        
        for _ in range(days):
            prediction = model.predict(current_price)
            predictions.append(prediction[0][0])
            current_price = prediction 
        
        predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))
        result_label.config(text=f"پیش‌بینی برای {days} روز آینده: {[int(p[0]) for p in predictions]}...")  
    except Exception as e:
        result_label.config(text=f"خطا: {e}")


win=tk.Tk()
win.geometry("800x600")
win.title("پیش بینی قیمت ")
win.config(bg="#007CAD")

tk.Label(win,text="تعداد روزای آینده رو وارد کن:" , font=("Arial" , 16),bg="#FFEE00").pack()

entry=tk.Entry(win,font=("Arial" , 16),bg="#7400C2",fg="#ffffff")
entry.pack()

tk.Button(win,text="پیش‌بینی" , command=predict_price, font=("Arial" , 16) , fg="#00C220",bg="#536600").pack()

result_label = tk.Label(win, text="",font=("Arial" , 16),bg="gray",fg="#ffffff")
result_label.pack()


win.mainloop()
```
